```python
# NUR ROHIM DEWANTARA
# 065002500012

def cek_prima(x):
    if x <= 1:
        return False
    if x == 2:
        return True
    for i in range(2, x):
        if x % i == 0:
            return False
    return True

def tampil_hasil_prima(angka, status_prima):
    if status_prima:
        print(f"{angka} adalah bilangan Prima")
    else:
        print(f"{angka} bukanlah bilangan Prima")

angka_input = int(input("Masukkan angka: ")
status = cek_prima(angka_input)
tampil_hasil_prima(angka_input, status)
```

```python
# NUR ROHIM DEWANTARA
# 065002500012

# ---  IMPLEMENTASI FUNGSI DENGAN PARAMETER ---
def cari_ordinal(parameter_angka):
    # Mengambil digit terakhir
    ekor = parameter_angka % 10

    if ekor == 1:
        return 'st'
    elif ekor == 2:
        return 'nd'
    elif ekor == 3:
        return 'rd'
    else:
        return 'th'

# ORDINAL NUMBER
print("Ordinal Number")
print("ketik 0 untuk mengentikan program")

while True:
    # --- 2. INPUT ---
    # Mengambil data dari pengguna
    angka_input = int(input("masukkan angka: "))

    # Memanggil fungsi (mengirim parameter)
    hasil_suffix = cari_ordinal(angka_input)

    # --- 3. OUTPUT ---
    # Menampilkan hasil dalam format tuple seperti di gambar
    print((angka_input, hasil_suffix))

    # Cek kondisi berhenti setelah output ditampilkan (agar 0 tetap muncul outputnya dulu)
    if angka_input == 0:
        print("terima kasih telah menggunakan program saya")
        break
```
