```python
# Nur Rohim Dewantara
# 065002500012
# Program Menentukan Jumlah Hari dalam Bulan

ulang = "y"

while ulang.lower() == "y":
    bulan = int(input("Masukkan nomor bulan (1-12): "))
    tahun = int(input("Masukkan tahun: "))

    if bulan in [1, 3, 5, 7, 8, 10, 12]:
        hari = 31
    elif bulan in [4, 6, 9, 11]:
        hari = 30
    elif bulan == 2:
        # Cek tahun kabisat
        if (tahun % 4 == 0 and tahun % 100 != 0) or (tahun % 400 == 0):
            hari = 29
        else:
            hari = 28
    else:
        print("Nomor bulan tidak valid!")
        hari = None

    if hari:
        print(f"Jumlah hari pada bulan {bulan} tahun {tahun} adalah {hari} hari.")

    ulang = input("Apakah ingin menghitung lagi? (y/n): ")

print("Program selesai.")
```
